### 3.Handle missing values

```python
df.isnull().sum()
```

```python
# Convert final_price to numeric and fill missing discount values
df['final_price'] = df['final_price'].astype(str).str.replace('"', '').str.replace('₹', '').str.replace(',', '').astype(float)
df['discount'] = df['discount'].fillna(0)
print("Data types after conversion and filling missing values:")
print(df[['final_price', 'discount']].dtypes)
print(df.isnull().sum())
```

```python
# Drop unnecessary columns
columns_to_drop = ['what_customers_said', 'seller_name', 'videos', 'seller_information', 'variations']
df = df.drop(columns=columns_to_drop)
print(f"Columns dropped. Remaining columns: {df.columns.tolist()}")
```

### 4.Filter products

```python
# filter rows
filtered_df = df[(df['rating'] > 4.5) & (df['discount'] > 70) & (df['final_price'] < 10000)]
display(filtered_df.head())
```

```python
# Select specific columns
columns_to_select = ['product_id', 'title', 'final_price', 'rating', 'category', 'initial_price', 'discount']
df_selected = df[columns_to_select]
print(f"Shape of selected DataFrame: {df_selected.shape}")
display(df_selected.head())
```

### 5. Handle duplicate records

```python
duplicates_removed = df.duplicated().sum()
if duplicates_removed > 0:
    df = df.drop_duplicates()
    print(f"Removed {duplicates_removed} duplicate rows.")
else:
    print("No duplicate rows found.")
print(f"Shape after handling duplicates: {df.shape}")
```

### 6. Create a new derived column

```python
# Since the dataset does not contain a quantity column,
# a new derived column 'potential_discount_amount' was created
# using the initial price and discount percentage
df['potential_discount_amount'] = df['initial_price'] * (df['discount'] / 100)
print("Added 'potential_discount_amount' column.")
display(df[['initial_price', 'discount', 'potential_discount_amount']].head())
```

### 7. Save the cleaned and processed dataset

```python
df.to_csv('cleaned_product_data.csv', index=False)
print("Cleaned data saved to 'cleaned_product_data.csv'")
```
